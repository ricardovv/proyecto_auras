## Prueba con imagen  

```python
# Abrir sesion
client = Client()
client.login('ricardov.bsky.social', 'santiago08')
```

```python
# cargar imagen, mostrrar c Matplotlib

l_imgs = [ 'imgs/img_adn_001.png','imgs/img_adn_002.png',
          'imgs/img_adn_003.png','imgs/img_adn_004.png'  ]
l_msjs = [ 'Boceto libro automatizado con datos de ADN',
           'Boceto para remix automatizado de "La Araucana" de Ercilla',
           'Construccin de grficos aleatorios' 
         ]

print( l_imgs[0] )
print( l_msjs[0] )
```

```python
# Postear con imagen - funciona

cual = 1xx

with open( l_imgs[ cual] , 'rb') as f:
    img_data = f.read()

    upload = client.upload_blob(img_data)
    images = [models.AppBskyEmbedImages.Image(alt='Img alt', image=upload.blob)]
    embed = models.AppBskyEmbedImages.Main(images=images)

    client.com.atproto.repo.create_record(
        models.ComAtprotoRepoCreateRecord.Data(
            repo=client.me.did,
            collection=models.ids.AppBskyFeedPost,
            record=models.AppBskyFeedPost.Record(
                created_at=client.get_current_time_iso(), text=l_msjs[ cual ], embed=embed
            ),
        )
    )

    ## of course, you can use the syntax sugar here instead
    #post = models.AppBskyFeedPost.Record(text='Text of the post', embed=embed, created_at=client.get_current_time_iso())
    #client.app.bsky.feed.post.create(client.me.did, post)
    ## or even high-level client
    #client.send_image(text='Text of the post', image=img_data, image_alt='Img alt')
    ## these three methods are equivalent
    
```
