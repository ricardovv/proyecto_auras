### A1- Scrapping: 

Web ejemplo:

https://www.surdoc.cl/registro/2-381


```python
##
#https://es.wikisource.org/wiki/La_araucana

l_ulr1 = [ 'https://es.wikisource.org/wiki/La_araucana_primera_parte/I',
          'https://es.wikisource.org/wiki/La_araucana_primera_parte/II'
        ]

cual = 1

## Preguntar por respuesta desde el servidor que aloja la página. 
req = requests.get(l_ulr1[cual])
larespuesta = req.status_code 

## mostrar respuesta, se puede demorar un poco
print( larespuesta )
```

```python
## Ver HTML  
soup = BeautifulSoup(req.content.decode("utf-8"),  "html.parser")
# mostrar el html
#soup

# Mostrar elemento a ver si toma
ejemplo = soup.find("h1").text.strip() 
# mostrar
print( ejemplo )
```

## Obtener elementos pagina

Cada canto son: 8 lineas x parrafo / 5 parrafos un bloque / 14 bloques 0 = 560 lineas


```python
# Ejemplos estáticos 
# obtiene el elemento 

# faltalimpiar nomre autor sin humero, y mencnion a la parte del titulo

#Titulo y autor
d_titulo_largo = soup.find("div", class_="middle-div").text.strip() 
d_titulo_largo = d_titulo_largo.partition("de ")
d_titulo = d_titulo_largo[0]
d_autor = d_titulo_largo[2]

# texto poema
d_num_canto = soup.find("div", class_="ws-div ws-h2-medio").text.strip() 
d_poema = soup.find( "div", class_="ws-bloque-centro ws-div" ).text.strip().split("\n")  
# largo poema
d_poema_len = len(d_poema)

# mosrtrar
print( d_titulo )
print( d_autor )
print( d_num_canto )
print( d_poema_len )
#print( d_poema )
d_poema
```

## Llevar a listas para crear DataFrame

```python
# crear lista vacias para llenar

l_titulo = []
l_autor  = []
l_poema  = []


# append elementos a listas segun largo de lista de url
for i in range( len(l_ulr1) ): # l_ulr1 
    req = requests.get(l_ulr1[i])
    larespuesta = req.status_code 
    soup = BeautifulSoup(req.content.decode("utf-8"),  "html.parser")

    #Titulo y autor
    d_titulo_largo = soup.find("div", class_="middle-div").text.strip() 
    d_titulo_largo = d_titulo_largo.partition("de ")
    d_titulo = d_titulo_largo[0]
    d_autor = d_titulo_largo[2]

    # texto poema
    d_num_canto = soup.find("div", class_="ws-div ws-h2-medio").text.strip() 
    d_poema = soup.find( "div", class_="ws-bloque-centro ws-div" ).text.strip().split("\n")  
    # largo poema
    d_poema_len = len(d_poema)

    
    l_titulo.append( d_titulo )
    l_autor.append( d_autor )
    l_poema.append( d_poema )


# Mostrar si los lee
print( l_titulo ) 
print( l_autor ) 
#print( l_poema ) 
l_poema
```

```python
# Crear dataframe c columnas planas definitivas

l_titulo_plano = []
l_autor_plano  = []
l_poema_plano  = []
cual = 0 

for i in range( len(l_poema[ cual ]) ): 
    cada_linea = l_poema[ cual ][i]
    print(i, l_titulo[cual], l_autor[cual], cada_linea )

    l_titulo_plano.append( l_titulo[cual] )
    l_autor_plano.append( l_autor[cual] )
    l_poema_plano.append( cada_linea )
```

## Guardar como csv 

```python
# dictionario de listas
dict_l_datos = {'titulo': l_titulo_plano, 'l_autor_plano': l_autor_plano, 'poema_linea': l_poema_plano}
    
df_araucana_lineas = pd.DataFrame( dict_l_datos )
     
# saving the dataframe
df_araucana_lineas.to_csv('df_araucana_lineas.csv')
```
